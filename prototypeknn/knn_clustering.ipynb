```python
# compute the embedding of each text.
import openai, numpy as np
from openai import OpenAI
client = OpenAI()

def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding

# Compute the embeddings of the summaries
df['embedding'] = df.ev_summary.apply(lambda x: get_embedding(x, model='text-embedding-3-small'))
```

```python
# Sort in place by 'Score'
df.sort_values(by='overall_score', inplace=True)
display(df)
```

```python
# run knn on the df
def cosine_distance(embedding1, embedding2):
    """
    Compute the cosine distance (1 - cosine similarity) between two vectors.
    """
    embedding1 = np.array(embedding1)
    embedding2 = np.array(embedding2)
    dot_product = np.dot(embedding1, embedding2)
    norm1 = np.linalg.norm(embedding1)
    norm2 = np.linalg.norm(embedding2)
    cosine_similarity = dot_product / (norm1 * norm2)
    return 1 - cosine_similarity

def find_group_if_valid(df, seed_idx, used_mask, k, distance_fn):
    """
    Find a group starting from the seed index if its k nearest neighbors are all unused.

    Args:
        df (pd.DataFrame): DataFrame with embeddings and used mask.
        seed_idx (int): Index of the seed text.
        used_mask (np.array): Boolean array marking used entries.
        k (int): Number of neighbors to find.
        distance_fn (callable): Custom distance function.

    Returns:
        list: Indices of the group if valid, otherwise an empty list.
    """
    # Compute distances to all unused embeddings
    seed_embedding = df.iloc[seed_idx]['embedding']
    distances = [
        (idx, distance_fn(seed_embedding, df.iloc[idx]['embedding']))
        for idx in range(len(df))
        if not used_mask[idx] and idx != seed_idx
    ]
    # Sort by distance and pick the top k closest
    distances = sorted(distances, key=lambda x: x[1])[:k]
    similar_indices = [idx for idx, _ in distances]

    # Validate if all k neighbors are unused
    if len(similar_indices) == k and all(not used_mask[idx] for idx in similar_indices):
        return [int(seed_idx)] + similar_indices  # Return group (seed + k neighbors)
    else:
        return []  # Invalid group


def form_groups_with_disjoint_nn(df, num_groups=3, group_size=5, distance_fn=cosine_distance):
    """
    Form disjoint groups ensuring all nearest neighbors of the seed are unused.
    """
    used_mask = np.zeros(len(df), dtype=bool)
    groups = []

    for _ in range(num_groups):
        # Step 1: Find the first unused seed text
        seed_candidates = np.where(~used_mask)[0]

        valid_group = []
        for seed_idx in seed_candidates:
            # Step 2: Attempt to form a valid group
            valid_group = find_group_if_valid(df, seed_idx, used_mask, k=group_size - 1, distance_fn=distance_fn)
            if valid_group:  # Found a valid group
                break

        if not valid_group:
            break  # No more valid groups can be formed

        # Step 3: Mark group members as used and add to groups
        for idx in valid_group:
            used_mask[idx] = True
        groups.append(valid_group)

    return groups


# Example usage
df = df.reset_index().drop(columns='index')

groups = form_groups_with_disjoint_nn(df, num_groups=10, group_size=3)
print("Groups:", groups)

def calculate_group_std(group, df):
    """
    Calculate the standard deviation of embeddings for a given group.
    
    Args:
        group (list): List of indices in the group.
        df (pd.DataFrame): DataFrame containing the 'embedding' column.
        
    Returns:
        float: Standard deviation of the embeddings for the group.
    """
    embeddings = np.array([df.iloc[idx]['embedding'] for idx in group])
    return np.std(embeddings, axis=0).mean()  # Compute mean of standard deviations along dimensions

# Calculate standard deviation for each group
group_std_devs = [ (i, calculate_group_std(group, df)) for i, group in enumerate(groups) ]

# Sort groups by standard deviation (ascending order to prioritize similarity)
sorted_groups = sorted(group_std_devs, key=lambda x: x[1])

# Reorder the groups based on the sorting
sorted_group_indices = [x[0] for x in sorted_groups]
sorted_groups_final = [groups[i] for i in sorted_group_indices]

# Initialize a dictionary to map indices to group numbers
index_to_group = {}

# Populate the dictionary with group numbers
for group_num, indices in enumerate(sorted_groups_final):
    for index in indices:
        index_to_group[index] = group_num

# Save csv with the groups
df['group'] = df.index.map(index_to_group).fillna(np.nan)
df.to_csv("/Users/mauroleidi/Desktop/HackaTUM24/HackaTUM24/prototypeknn/df_with_embedding_and_sorted_groups.csv")
```
